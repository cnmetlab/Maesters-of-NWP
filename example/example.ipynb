# Get variable dataset predicted at hours from batch date time

```python
from maesters import Maester
from matplotlib import pyplot as plt

enfo = Maester('ecmwf','enfo',batch ='2022-07-02 00:00',hour=[3,27],varname='TP_SFC').xarray()
icon = Maester('dwd','icon',batch='2022-07-02 00:00',hour=[3,27],varname='TP_SFC').xarray()
geps = Maester('cmc','geps_ens',batch='2022-07-02 00:00',hour=[3,27],varname='TP_SFC').xarray()
oper = Maester('ecmwf','oper',batch='2022-07-02 00:00',hour=[3,27],varname='TP_SFC').xarray()

# plot 24 hours accumulated precipitation from different sources
fig,axs = plt.subplots(2,2,figsize=(12,10))
(enfo['TP_SFC'].diff('time').sel(lon=slice(70,140),lat=slice(60,15))).plot(vmin=0,vmax=0.2,ax=axs[0,0])
(oper['TP_SFC'].diff('time').sel(lon=slice(70,140),lat=slice(60,15))).plot(vmin=0,vmax=0.2,ax=axs[0,1])
(icon['TP_SFC'].diff('time').sel(lon=slice(70,140),lat=slice(15,60))/1000).plot(vmin=0,vmax=0.2,ax=axs[1,1])
(geps['TP_SFC'].diff('time').sel(lon=slice(70,140),lat=slice(15,60))/1000).plot(vmin=0,vmax=0.2,ax=axs[1,0])
```
